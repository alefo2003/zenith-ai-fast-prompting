```python
# Definimos un pequeño dataset (Mock Data) con atributos estéticos y técnicos
# Esto demuestra que no dependemos de tags genéricos, sino de metadatos profundos.

anime_catalogue = [
    {
        "titulo": "Cowboy Bebop",
        "estilo_visual": "Retro 90s, granulado cinematográfico, paleta de colores urbana",
        "ritmo": "Episódico, pausado con explosiones de acción",
        "estudio": "Sunrise",
        "esencia": "Melancolía espacial, jazz, soledad, existencialismo"
    },
    {
        "titulo": "Mushishi",
        "estilo_visual": "Acuarela, fondos orgánicos, naturaleza detallada",
        "ritmo": "Lento, contemplativo, atmosférico",
        "estudio": "Artland",
        "esencia": "Zen, folclore japonés, calma, misterio natural"
    },
    {
        "titulo": "Cyberpunk: Edgerunners",
        "estilo_visual": "Colores neón vibrantes, animación frenética, alto contraste",
        "ritmo": "Rápido, caótico, intenso",
        "estudio": "Trigger",
        "esencia": "Distopía, rebeldía, violencia estilizada, tecnología"
    },
    {
        "titulo": "Violet Evergarden",
        "estilo_visual": "Iluminación volumétrica, ultra-detallado, paleta pastel y oro",
        "ritmo": "Dramático, pausado",
        "estudio": "Kyoto Animation",
        "esencia": "Belleza emocional, superación, época victoriana"
    }
]

print(f"✅ Catálogo de {len(anime_catalogue)} series cargado con éxito.")
```

```python
# Celda 4: Definición de la función
def ani_mind_engine(user_query):
    # Usamos el catálogo que definiste en la Celda 3
    prompt = f"""
    Eres 'AniMind', un experto en estética de anime.
    Basándote SOLO en este catálogo: {anime_catalogue}

    Analiza la petición del usuario: "{user_query}"

    Responde en formato Markdown:
    1. Anime recomendado.
    2. Por qué coincide con la 'esencia' buscada.
    3. Un 'Stylized Prompt' técnico para generar una imagen de este anime.
    """

    try:
        response = model.generate_content(prompt)
        display(Markdown(response.text))
    except Exception as e:
        print(f"❌ Error al generar contenido: {e}")

print("✅ Función AniMind lista.")
```

```python
# Celda Interactiva
print("--- BIENVENIDO A ANIMIND ---")
pregunta = input("¿Qué tipo de atmósfera buscas hoy? ")

if pregunta:
    print("\nBuscando en la base de datos de esencia...\n")
    ani_mind_engine(pregunta)
```
